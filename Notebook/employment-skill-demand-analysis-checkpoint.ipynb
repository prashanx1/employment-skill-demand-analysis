```python
import pandas as pd

jobs = pd.read_csv(r"C:\Users\prash\Downloads\archive\DataAnalyst.csv")

print(jobs.head())
print(jobs.columns)
```

```python
print(jobs.columns)
```

```python
jobs = jobs[
    ['Job Title', 'Company Name', 'Location', 'Salary Estimate', 'Job Description']
]

jobs.columns = [
    'job_title',
    'company',
    'location',
    'salary',
    'job_description'
]

jobs.head()
```

```python
jobs = jobs.copy()
```

```python
jobs.loc[:, 'job_title'] = jobs['job_title'].str.lower().str.strip()
jobs.loc[:, 'company'] = jobs['company'].str.lower().str.strip()
jobs.loc[:, 'location'] = jobs['location'].str.lower().str.strip()

jobs = jobs.dropna(subset=['job_description'])
```

```python
skills = [
    'python', 'sql', 'excel', 'power bi', 'tableau',
    'machine learning', 'statistics', 'data analysis',
    'communication', 'java'
]

def extract_skills(text):
    found = []
    for skill in skills:
        if skill in text.lower():
            found.append(skill)
    return ', '.join(found)

jobs['skills_required'] = jobs['job_description'].apply(extract_skills)

jobs[['job_title', 'skills_required']].head()
```

```python
jobs.to_csv("jobs_cleaned_with_skills.csv", index=False)
```

```python
import matplotlib.pyplot as plt

top_locations = jobs['location'].value_counts().head(10)

plt.figure()
top_locations.plot(kind='bar')
plt.title("Top Job Locations")
plt.xlabel("Location")
plt.ylabel("Number of Jobs")
plt.tight_layout()
plt.show()
```

```python
skill_series = jobs['skills_required'].str.split(', ')
all_skills = skill_series.explode()

top_skills = all_skills.value_counts().head(10)

plt.figure()
top_skills.plot(kind='bar')
plt.title("Top Required Skills")
plt.xlabel("Skill")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
